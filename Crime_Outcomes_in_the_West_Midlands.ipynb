```python
import pandas as pd
import glob

files = glob.glob('/content/West Mid Police Data/*.csv')
df = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

df["Last outcome category"].value_counts()
```

```python
# Checking for missing values

df["Last outcome category"].isna().sum()
```

```python
df["Last outcome category"] = df["Last outcome category"].fillna("Outcome unknown")
```

```python
outcome_counts = df["Last outcome category"].value_counts()
outcome_percent = df["Last outcome category"].value_counts(normalize=True) * 100

print("\nOutcome Percentages:")
display(outcome_percent.round(2))
```

```python
import matplotlib.pyplot as plt

outcome_counts.plot(kind="bar")

plt.title("Crime Investigation Outcomes")
plt.ylabel("Number of Cases")
plt.xticks(rotation=90)

plt.show()
```
